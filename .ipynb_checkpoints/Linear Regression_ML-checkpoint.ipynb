## Data preprocessing

 Convert categorical variables to numeric using one-hot encoding

```python
def load_data():
    # Load the dataset
    df = pd.read_csv('Medical Price Dataset.csv')
    df = pd.get_dummies(df, columns=['smoker', 'sex', 'region'])

    # Separate the charge from the features
    x_train = df.drop('charges', axis=1).values
    y_train = df['charges'].values
    
    return x_train, y_train
```

```python
def linear_regression(x_train, y_train):
    # Add bias term to x_train
    x_train = np.insert(x_train, 0, 1, axis=1)
    
    # Calculate the weights using normal equation
    x_transpose = np.transpose(x_train)
    x_transpose_dot_x = np.dot(x_transpose, x_train)
    x_transpose_dot_y = np.dot(x_transpose, y_train)
    weights = np.linalg.solve(x_transpose_dot_x, x_transpose_dot_y)
    
    return weights
```

```python
def predict(x_test, weights):
    #Bias
    x_test = np.insert(x_test, 0, 1, axis=1)
    predictions = np.dot(x_test, weights)
    
    return predictions
```

```python
def visualize(x_train, y_train, weights):
    #Bias
    x_train = np.insert(x_train, 0, 1, axis=1)
    
    # Calculate predictions
    predictions = np.dot(x_train, weights)
    
    # Plot actual vs predicted values
    plt.scatter(range(len(y_train)), y_train, color='blue', label='Actual')
    plt.plot(range(len(predictions)), predictions, color='red', label='Predicted')
    plt.xlabel('Sample')
    plt.ylabel('Medical Cost')
    plt.title('Linear Regression')
    plt.legend()
    plt.show()
```

```python
def calculate_r_squared(y_actual, y_predicted):
    y_mean = np.mean(y_actual)
    ss_total = np.sum((y_actual - y_mean) ** 2)
    ss_residual = np.sum((y_actual - y_predicted) ** 2)
    r_squared = 1 - (ss_residual / ss_total)
    return r_squared
def calculate_mse(y_actual, y_predicted):
    mse = np.mean((y_actual - y_predicted) ** 2)
    return mse
```

## Accuracy 
using MSE nad R square value

```python
# Load and preprocess the data
x, y = load_data()

# Perform train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Perform linear regression
weights = linear_regression(x_train, y_train)

# Make predictions on the training set
train_predictions = predict(x_train, weights)

# Calculate MSE on the training set
train_mse = calculate_mse(y_train, train_predictions)
print("Train Mean Squared Error (MSE): {:.2f}".format(train_mse))

# Perform linear regression
weights = linear_regression(x_test, y_test)

# Make predictions on the test set
test_predictions = predict(x_test, weights)

# Calculate MSE on the test set
test_mse = calculate_mse(y_test, test_predictions)
print("Test Mean Squared Error (MSE): {:.2f}".format(test_mse))

# Visualize the results
visualize(x_test, y_test, weights)
```

```python
# Calculate R-squared value
r_squared = calculate_r_squared(y_train, train_predictions)

print("R-squared value: {:.2f}".format(r_squared))
```
